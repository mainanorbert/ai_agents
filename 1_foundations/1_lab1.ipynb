```python
import anthropic
load_dotenv(override=True)


anthropic_api_key = os.getenv('ANTHROPIC_API_KEY')

if anthropic_api_key:
    print(f"Anthropic API Key exists and begins {anthropic_api_key[:8]}")
else:
    print("Anthropic API Key not set - please head to the troubleshooting guide in the setup folder")
```

```python
claude_client = anthropic.Anthropic(api_key=anthropic_api_key)



def select_an_area_of_interest():
    messages = [
        {"role": "user", "content": """"Please select any broad business area that is commercially viable for an AI applications.
        
        Respond only with one clear the area of interest without descriptions.
        """}
    ]
    response = claude_client.messages.create(
    model="claude-haiku-4-5",
    max_tokens=200,
    temperature=0.7,
    messages=messages)
    return response.content[0].text
```

```python
print(select_an_area_of_interest())
```

```python
def agent2_idea_generator(area):
    messages = [
        {"role": "user", "content": f"""
        You are an AI agent that generates ideas for a business.
        The business area is {area}.
        Generate one clear pinpoint idea in {area}  that can be implemented using AI agentic AI to solve a challenging problem. Responde only in 1 line.
        """}
    ]

    response = openai.chat.completions.create(
        model="gpt-4.1-mini",
        messages=messages
    )
    return response.choices[0].message.content
```

```python
area = select_an_area_of_interest()
print(agent2_idea_generator(area))
```

```python
def agent3_solution_generator(idea):
    messages = [
        {"role": "user", "content": f"""
        You are an AI agent that generates solutions for a business.
        The business idea is {idea}.
        Provide a clear AI agent-based solution.
        Include on this:
        - Architecture
        - Tools/tech stack
        - Step-by-step workflow

        Be precise and concise.
        """}
    ]

    response = openai.chat.completions.create(
        model="gpt-4.1-mini",
        messages=messages
    )
    return response.choices[0].message.content
```

```python
area = select_an_area_of_interest()
idea = agent2_idea_generator(area)
solution = agent3_solution_generator(idea)


print(display(Markdown(solution)))
```
